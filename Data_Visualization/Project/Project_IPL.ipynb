# Key player performance

1. most 'Player of the match' awards

```python
award = df['player_of_the_match'].value_counts().head(10)
sns.barplot(y = award.index, x = award.values, palette='rainbow')
```

```python
# 2. Top Scorers

scorer = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
```

```python
scorer
```

```python
scorer.plot(kind = 'barh')
```

```python
# 3. Best Bowling figures

df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x:x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head()
top_bowlers.plot(kind = 'barh')
```

```python
top_bowlers
```

```python
# 4. Most matches Plyed by venue

venue = df['venue'].value_counts()
venue
```

```python
sns.barplot(y = venue.index, x = venue.values, hue=venue.index)
```

```python
# Q1 : Who won the highest margin by runs

df[df['won_by'] == 'Runs'].sort_values(by = "margin", ascending=False).head(1)[['match_winner','margin']]
```
