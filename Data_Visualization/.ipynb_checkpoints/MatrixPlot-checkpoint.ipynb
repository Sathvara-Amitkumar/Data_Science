```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Heat Map**

```python
flights = sns.load_dataset("flights")
flights
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tipscorr = tips[['total_bill','tip','size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(), annot=True)
```

# **Cluster Map**

```python
sns.clustermap(tipscorr.corr())
```

# **pivot Table Heat Map**

```python
pvflights = flights.pivot_table(values='passengers', index='month', columns='year')
```

```python
pvflights
```

```python
sns.heatmap(pvflights.corr(), annot=True)
```

# **Regression Plot**

```python
sns.lmplot(x = 'total_bill', y='tip', data=tips, hue='sex', scatter_kws={'s':40}, markers=['o','v'])
```

```python
sns.lmplot(x = 'total_bill', y='tip', data=tips, hue='sex', scatter_kws={"s": 30, "color": "red"}, height=4)
```
