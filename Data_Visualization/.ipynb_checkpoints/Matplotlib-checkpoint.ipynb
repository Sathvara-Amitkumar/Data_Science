```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
x
y = x ** 2
y
```

```python
plt.plot(x,y)
plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x**2, x**3)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x**3, x**3)
```

**Object Oriented Way**

```python
fig = plt.figure(figsize=(10,8), dpi=100)
axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])
axis2.plot(y,x)
```
