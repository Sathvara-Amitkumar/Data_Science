```python
def extract_eps(txt):
    check = False
    data = ""
    
    for i in txt:
        if i == ")":
            check = False
            return data
        if check == True:
            data += i
        if i == "(":
            check = True
    return data
```

```python
df["Episodes"] = df["Title"].apply(extract_eps)
```

```python
df["Episodes"] = df["Episodes"].str.replace(" eps", "")
```

```python
df["Episodes"] = df["Episodes"].astype(int)
```

```python
# def timestamp(txt):
#     check = False
#     time = ""
#     for i in txt:
#         if check == True:
#             if i == ",":
#                 check = False
#                 break
#             else:
#                 if len(time) != 19:
#                     time += i
#         if i == ")":
#             check = True
#     return time

def extract_time(txt):
    check = False
    data = ""
    for i in range(len(txt)):
        if txt[i] == ")":
            for j in range(i+1, i+20):
                data += txt[j]

            return data
```

```python
df["Total Time"] = df['Title'].apply(extract_time)
```

```python
df.drop(columns="Timestamps", inplace=True)
```

```python
df.head()
```

```python
def calculate_total_months(time_str):
    try:
        start_str, end_str = time_str.split(" - ")
        
        start = pd.to_datetime(start_str, format="%b %Y")
        end = pd.to_datetime(end_str, format="%b %Y")
        
        # Difference in months (+1 for inclusive count)
        total_months = (end.year - start.year) * 12 + (end.month - start.month) + 1
        return total_months
    except Exception:
        return None 
        
df["Total Months"] = df["Total Time"].apply(calculate_total_months)
```
